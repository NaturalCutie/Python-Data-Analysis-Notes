项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

导入NumPy和Pandas数据库，用Pandas的`read_csv`函数读取原始数据集'e_commerce.csv'，使其转换成DataFrame格式，并赋值给变量`df`。

展示数据集的前5行和后5行。

## 评估数据

在这一部分，我将对上一部分所建立的`df`这个DataFrame所包含的数据进行评估

评估主要从两个方面进行：结构和内容，即整齐度和干净度。数据的结构性问题指不符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”这三个标准，数据的内容性问题包括存在丢失数据、重复数据、无效数据等。

### 一、评估数据整洁度

利用DataFrame的info方法和sample方法，分别提供数据的概况信息和实际数据，来评估数据的整洁度

```python
df.info()
```

列名符合每列只能表示一个变量的整洁性规则

```python
df.sample(5)
```

从抽样的10行数据来看，符合'每列是一个变量，每行是一个观察值，每个单元格是一个值'这三个标准。具体来看每行是关于某商品的一次交易数据，每列是关于交易的各个变量，因此不存在结构性问题。

### 二、评估数据干净度

```python
df.info()
```

从输出结果来看，数据共有541909条观察值。

`Description`和`CustomerID`列存在缺失值。

`CustomerID`列数据类型应为object，`InvoiceDate`数据类型应为datetime，应当进行数据类型转换

#### (一)、评估丢失数据

初步统计各列缺失值的数量

```python
df.isnull().sum()
```

在了解`Description`存在缺失值后，根据条件提取出缺失观察值。

```python
df[df.Description.isnull()]
```

有1454条交易数据缺失`Description`变量值

从输出结果来看，这些缺失`Description`的交易数据，`UnitPrice`都为0，且`CustomerID`同时缺失。为了验证猜想，我们增加筛选条件，看是否存在`Description`变量缺失，且`UnitPrice`不为0的数据

```python
df[(df.Description.isnull()) & (df.UnitPrice != 0)]
```

验证猜想，输出结果为0条，说明缺失`Description`变量值的数据，同时也不具备有效的`UnitPrice`值。

`Description`表示产品名称，`UnitPrice`表示产品单价，都是进行后续数据分析的关键变量。如果他们同时缺失，我们认为这些数据无法提供有效含义，因此这些后续可以被删除。

变量`CustomerID`同样存在缺失值，因此也根据条件提取出缺失观察值。

```python
df[df.CustomerID.isnull()]
```

有135080条交易数据缺失`CustomerID`变量值

从输出结果来看，缺失`CustomerID`变量值不一定缺失其他变量值，并且`CustomerID`不是后续分析畅销商品的关键变量，因此保留此变量为空的观察值。

#### (二)、评估重复数据

根据数据变量含义来看，虽然`InvoiceNo`、`StockCode`和`CustomerID`都是唯一标识符，但是一次交易可以包含多件商品，因此`InvoiceNo`可以存在重复；不同交易可以选择同件商品，因此`InvoiceNo`可以存在重复；每个客户可以产生不同交易，因此`CustomerID`可以存在重复。

但是数据所有变量同时重复时，应被看作重复数据。而不是某个顾客一次买了多件同一商品，毕竟这样应该是quantity变量的值为多个。

```python
df[df.duplicated()]
```

从输出结果来看，数据中存在5268条重复数据，应在数据清洗环节删除所有变量同时重复的数据。

#### (三)、 评估不一致数据 

从数据变量含义来看，只能判断`Country`变量是否存在不一致数据

```python
df.Country.value_counts()
```

从`Country`变量值来看，此处U.K., UK和United Kingdom都表示英国，United States和USA都表示美国。这些数据是不一致数据，在后续清理步骤需要把数值都替换成统一的

#### (四)、评估无效/错误数据

通过DataFrame的`describe`方法，对数值统计信息进行快速了解

```python
df.describe()
```

从输出结果来看，`Quantity`和`UnitPrice`的值存在负数，这些均为无效数据。

接下来先筛选出`Quantity`为负数的观察值，来进一步观察数据的特点

```python
df[df['Quantity'] <= 0]
```

可以发现`Quantity`为负数时，似乎`InvoiceNo`均为C开头，表示这笔交易被取消。

为了验证猜想，我们需要增加筛选条件，看是否存在`Quantity`变量为负数且`InvoiceNo`不以C开头的观察值。

```python
df[(df['Quantity'] <= 0) & (df['InvoiceNo'].str[0] != 'C')]
```

可以发现猜想并不准确，确实存在Quantity变量为负数且InvoiceNo不以C开头的观察值。

但进一步观察发现，这些筛选出的观察值UnitPrice均为0，因此我们增加筛选条件，进一步验证

```python
df[(df['Quantity'] <= 0) & (df['InvoiceNo'].str[0] != 'C') & (df['UnitPrice'] != 0)]
```

根据输出结果，猜想得到验证，证明当`Quantity`为负数的时候，观察值满足以下条件之一：

`InvoiceNo`以C开头，表示订单被取消；

`UnitPrice`为0。说明单价0英镑

这些交易数据不是有效交易数据，不在后续数据分析范围内，因此在数据清理阶段，将`Quantity`为负数的观察值删除

同时，我们需要确认，`InvoiceNo`以C开头时，或`UnitPrice`为0时，`Quantity`变量是否都为负数。

这样可以帮我们确认，在数据清理环节，是否只需要删除`Quantity`为负数的数据

```python
df[(df['InvoiceNo'].str[0] == 'C') & (df['Quantity'] > 0)]
```

根据输出结果，`InvoiceNo`以C开头时，`Quantity`变量的值都大于0。因此在数据清理阶段，将`Quantity`为负数的观察值删除，就可以删除所有订单被取消的数据。

因为在评估缺失数据时，发现了一批`Description`为空值，同时`UnitPrice`为0的数据。因此在筛选时还要加上条件，`Description`不为空值

```python
df[(df['UnitPrice'] == 0) & (df['Quantity'] > 0) & (df['Description'].notnull())]
```

根据输出结果，存在`UnitPrice`为0时，`Quantity`大于0，且`Description`不为空值的数据。

但进一步观察发现，这些筛选出的观察值，大部分`CustomerID`为NaN，因此我们增加筛选条件，进一步验证

```python
df[(df['UnitPrice'] == 0) & (df['Quantity'] > 0) & (df['Description'].notnull()) & (df['CustomerID'].notnull())]
```

根据输出结果，猜想得到验证，证明当`UnitPrice`为0时，`Quantity`大于0，且`Description`不为空值时的数据，大部分`CustomerID`为空值，小部分不为空值。

但由于数据分析的最终目的是，提高营收，因此在数据清理环节，删除`UnitPrice`为0的数据。

接下来筛选出`UnitPrice`为负数的观察值，观察数据的特点

```python
df[df['UnitPrice'] < 0]
```

从输出结果来看，`UnitPrice`为负数的观察值都是坏账调整，不属于实际交易的数据，因此在数据清洗环节也将其删除

通过DataFrame的`value_counts`方法，分别检查`InvoiceNo`、`StockCode`、`InvoiceDate`变量是否存在可被发现的无效/错误数据

```python
df.InvoiceNo.value_counts()
```

```python
df.StockCode.value_counts()
```

```python
df.InvoiceDate.value_counts()
```

从输出结果来看，不存在未被发现过的无效/错误数据。

## 清理数据

根据前面数据评估部分的结果，我们需要进行的数据清理包括：

- 把`CustomerID`变量的数据类型转换为字符串类型
- 把`InvoiceDate`变量的数据类型转换为日期时间类型
- 把`Description`变量存在缺失的观察值删除
- 删除所有变量同时重复的数据
- 把`Country`变量值`'U.K.'`、`'UK'`替换为`'United Kindom'`
- 把`Country`变量值`'USA'`替换为`'United States'`
- 把`Quantity`小于等于0的观察值删除
- 把`UnitPrice`观察值小于等于0的数据删除

为了区分经过清理的数据和原始的数据，我们创建新的变量`cleaned_df`。之后的清理步骤都将被运用在`cleaned_df`上。

```python
cleaned_df = df
cleaned_df
```

把`CustomerID`变量的数据类型转换为字符串类型：

```python
cleaned_df['CustomerID'] = cleaned_df['CustomerID'].astype(str)
cleaned_df['CustomerID']
```

把`CustomerID`结尾的`".0"`删除：

```python
cleaned_df['CustomerID'] = cleaned_df['CustomerID'].str.slice(0, -2)
cleaned_df['CustomerID']
```

把`InvoiceDate`变量的数据类型转换为日期时间类型：

```python
cleaned_df['InvoiceDate'] = pd.to_datetime(cleaned_df['InvoiceDate'])
cleaned_df.InvoiceDate
```

把`Description`变量存在缺失的观察值删除：

```python
cleaned_df.dropna(subset=['Description'], inplace=True)
cleaned_df.Description.isnull().sum()
```

删除所有变量同时重复的数据：

```python
cleaned_df.drop_duplicates(inplace=True)
cleaned_df[cleaned_df.duplicated()]
```

把`Country`变量值`'U.K.'`、`'UK'`替换为`'United Kindom'`;
把`Country`变量值`'USA'`替换为`'United States'`：

```python
cleaned_df.replace(['U.K.', 'UK'], 'United Kingdom', inplace=True)
cleaned_df.replace('USA', 'United States', inplace=True)
print(len(cleaned_df[cleaned_df['Country'] == 'U.K.']))
print(len(cleaned_df[cleaned_df['Country'] == 'UK']))
print(len(cleaned_df[cleaned_df['Country'] == 'USA']))
```

把`Quantity`小于等于0的观察值删除：

```python
cleaned_df.drop(cleaned_df[cleaned_df.Quantity <= 0].index, inplace=True)
len(cleaned_df[cleaned_df.Quantity <= 0])
```

把`UnitPrice`观察值小于等于0的数据删除：

```python
cleaned_df.drop(cleaned_df[cleaned_df.UnitPrice <= 0].index, inplace=True)
len(cleaned_df[cleaned_df.UnitPrice <= 0])
```

发现单价存在0.001英镑的，不合常理，应为错误数据

```python
cleaned_df.drop(cleaned_df[cleaned_df.UnitPrice == 0.001].index, inplace=True)
cleaned_df.UnitPrice.sort_values()
```
